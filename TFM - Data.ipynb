# Generar los factores

```python
def metrica_tickers(fundamentals, metrica):
    raw = fundamentals[fundamentals['dimension']=='ART'][[metrica,'ticker','calendardate']]
    raw.rename(columns={'calendardate':'date'}, inplace=True)
    raw = raw.dropna()
    metrica_discrete = raw.pivot_table(index=['date'], columns='ticker', values=metrica)
    
    full_history_metrica = pd.DataFrame(index=pd.date_range(metrica_discrete.index.min(), metrica_discrete.index.max()))
    full_history_metrica.index.name = 'date'

    full_history_metrica = full_history_metrica.merge(metrica_discrete, how='left', 
                                              left_index=True, right_index=True
                                             ).fillna(method='ffill')
    return full_history_metrica

def metrica_daily_tickers(daily, metrica):
    raw = daily[['date','ticker',metrica]].dropna()
    raw = raw.dropna()
    metrica_discrete = raw.pivot_table(index=['date'], columns='ticker', values=metrica)
    
    full_history_metrica = pd.DataFrame(index=pd.date_range(metrica_discrete.index.min(), metrica_discrete.index.max()))
    full_history_metrica.index.name = 'date'

    full_history_metrica = full_history_metrica.merge(metrica_discrete, how='left', 
                                              left_index=True, right_index=True
                                             ).fillna(method='ffill')
    return full_history_metrica
```

## Momentum

```python
def momentum(pricing, offset=253):
    return (pricing/pricing.shift(offset)-1)

def compute_future_returns(pricing, offset=253):
    return (pricing.shift(-offset)/pricing-1)
```

# $\beta$ mercado

$\beta$ es la sensibilidad del precio de un activo al mercado en su conjunto. Para calcularla podemos utilizar el hist√≥rico de precios del SP500 y de la acci√≥n de la que queremos extraer la $\beta$ mediante una regresi√≥n o las varianzas/covarianzas.

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    
def compute_beta(date, ticker, SP_daily_data, pricing):
    end_date = date
    init_date = end_date - pd.tseries.offsets.DateOffset(years=3)
    # Market proxy for serie
    print("Compute beta of {} in {}".format(ticker, end_date))
    SP500_proxy = SP_daily_data.loc[init_date:end_date]
    SP500_proxy.index.name='date'
    # Series of historical data of the stcok
    stock_series = pricing[[ticker]].loc[init_date:end_date]
    
    stocks_vs_market = SP500_proxy.merge(stock_series, how='inner', right_index=True, left_index=True)
    stocks_vs_market=stocks_vs_market.resample('BM').apply(lambda x: x[-1]).pct_change()[1:]
    
    print("Using data with shape {}".format(stocks_vs_market.shape))
    cov_matrix = stocks_vs_market.cov()
    beta = cov_matrix.loc['Value'][ticker]/cov_matrix.loc['Value']['Value']
    #beta using linear regression
    try:
        reg = LinearRegression().fit(stocks_vs_market[ticker].values.reshape(-1,1),
                                 stocks_vs_market['Value'].values.reshape(-1,1))
    
        #plt.scatter(x=stocks_vs_market[ticker].values.reshape(-1,1),
        #    y=stocks_vs_market['Value'].reshape(-1,1))
        #abline(reg.coef_[0], reg.intercept_)
        #plt.plot()
        beta_reg = reg.coef_[0][0]
    except:
        print("Not enought data to compute regression beta")
        beta_reg = float('NaN')
        
    return beta, beta_reg
```

```python
date = pd.to_datetime('2018-10-10')
compute_beta(date, "AAL", SP500_data, pricing)
```

# Unir los factores en un dataframe

```python
def build_factor_df(factor, name):
    factor_df = pd.DataFrame(factor.stack())
    factor_df.columns = [name]
    factor_df.index.set_names(['date','ticker'], inplace=True)
    return factor_df
```

```python
momentum_6m_f = build_factor_df(momentum_6m, 'momentum_6m')
momentum_11m_f = build_factor_df(momentum_11m, 'momentum_11m')
roe_f = build_factor_df(roe, 'roe')
beta_3Y_f = build_factor_df(beta_3Y, 'beta_3Y')
beta_3Y_reg_f = build_factor_df(beta_3Y_reg, 'beta_3Y_coef')
pb_f = build_factor_df(pb, 'price-to-book')
pe_f = build_factor_df(pe, 'earnings-to-price')
ps_f = build_factor_df(ps, 'price-to-sales')
market_cap_f = build_factor_df(market_cap, 'marketcap')
ev_f = build_factor_df(ev, 'enterprise-value')
evebit_f = build_factor_df(evebit, 'evebit')
evebitda_f = build_factor_df(evebit, 'evebitda')
returns_12m_f = build_factor_df(returns_12m, 'returns_12m')
returns_12m_lagged_12m_f = build_factor_df(returns_12m_lagged_12m, 'returns_12m_lagged_12m')
returns_12m_lagged_24m_f = build_factor_df(returns_12m_lagged_24m, 'returns_12m_lagged_24m')
opcash2price_f = build_factor_df(opcash2price, 'operating cashflow-to-price')
invest2price_f = build_factor_df(invest2price, 'investment-to-price')
eps_f = build_factor_df(eps, 'earnings-per-share')
current_ratio_f = build_factor_df(current_ratio, 'current-ratio')
opcash2equity_f = build_factor_df(opcash2equity, 'operating cashflow-to-equity')
capex_f = build_factor_df(capex, 'capex')
```

```python
factors_arr = [momentum_6m_f, momentum_11m_f,
              roe_f, beta_3Y_f, beta_3Y_reg_f,
               pb_f, pe_f,
              ps_f, returns_12m_f, market_cap_f,
              ev_f, evebit_f, evebitda_f,
              returns_12m_lagged_12m_f, returns_12m_lagged_24m_f,
              opcash2price_f, invest2price_f, eps_f,
              current_ratio_f, opcash2equity_f, capex_f]
```

```python
all_factors_merged = factors_arr[0]
for i in range(1,len(factors_arr)):
    all_factors_merged = all_factors_merged.merge(factors_arr[i], how='outer', left_index=True, right_index=True)
```

```python
all_factors = all_factors_merged
```
